```python
df['층'] = df['층'].astype(int)
```

```python
dfX = df[['전용면적', '층', '건축년도', '계약년', '계약월',
       '주택담보대출', '미국달러', '일본엔', '유로', '소비자물가지수',
       'KOSPI_종가', 'KOSDAQ_종가', '무담보콜금리', 'KORIBOR', 'CD',
       '국고채_1년', '국고채_10년']]
dfy = df['면적당가격']
```

```python
import statsmodels.api as sm
model = sm.OLS(dfy, dfX)
result = model.fit()
result.summary()
```
