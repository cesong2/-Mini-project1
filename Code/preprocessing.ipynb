```python
# 컬럼명
df1 = df1.rename(columns = df1.iloc[15])
df2 = df2.rename(columns = df2.iloc[15])
df3 = df3.rename(columns = df3.iloc[15])
df4 = df4.rename(columns = df4.iloc[15])
df5 = df5.rename(columns = df5.iloc[15])
df6 = df6.rename(columns = df6.iloc[15])
```

```python
# 불필요 행 삭제
df1 = df1.drop(range(16))
df2 = df2.drop(range(16))
df3 = df3.drop(range(16))
df4 = df4.drop(range(16))
df5 = df5.drop(range(16))
df6 = df6.drop(range(16))
```

```python
# 필요한 컬럼만 따오기
df1 = df1[['시군구','전용면적(㎡)','계약년월','계약일','거래금액(만원)','층','건축년도']]
df2 = df2[['시군구','전용면적(㎡)','계약년월','계약일','거래금액(만원)','층','건축년도']]
df3 = df3[['시군구','전용면적(㎡)','계약년월','계약일','거래금액(만원)','층','건축년도']]
df4 = df4[['시군구','전용면적(㎡)','계약년월','계약일','거래금액(만원)','층','건축년도']]
df5 = df5[['시군구','전용면적(㎡)','계약년월','계약일','거래금액(만원)','층','건축년도']]
df6 = df6[['시군구','전용면적(㎡)','계약년월','계약일','거래금액(만원)','층','건축년도']]
```

```python
# 계약년월일로 합치기
df1['계약년월일'] = df1['계약년월'] + df1['계약일'].apply(lambda x : '0'+x if len(x)==1 else x)
df2['계약년월일'] = df2['계약년월'] + df2['계약일'].apply(lambda x : '0'+x if len(x)==1 else x)
df3['계약년월일'] = df3['계약년월'] + df3['계약일'].apply(lambda x : '0'+x if len(x)==1 else x)
df4['계약년월일'] = df4['계약년월'] + df4['계약일'].apply(lambda x : '0'+x if len(x)==1 else x)
df5['계약년월일'] = df5['계약년월'] + df5['계약일'].apply(lambda x : '0'+x if len(x)==1 else x)
df6['계약년월일'] = df6['계약년월'] + df6['계약일'].apply(lambda x : '0'+x if len(x)==1 else x)
```

```python
# 데이터 합치기
df = pd.concat([df1, df2, df3, df4, df5, df6])
```

```python
df = df[['시군구','전용면적(㎡)','거래금액(만원)','층','건축년도','계약년월일']]
```

```python
df2 = df                                                # 17년 이후 자료
df1 = pd.read_csv('C:/부동산/CGKim/data/houseprice.csv') # 17년 이전 자료
```

```python
df1 = df1[['시군구','전용면적(㎡)','거래금액(만원)','층','건축년도','계약년월일']]
```

```python
df = pd.concat([df1, df2])
df
```

```python
df['계약년월일'] = df['계약년월일'].astype(str)
```

```python
# 시간대별로 정리
df = df.sort_values('계약년월일')
```

```python
# 결측치 확인
df.info()
```

```python
# 결측치 2행 제거
df.dropna(inplace=True)
```

```python
# 숫자 사이의 ','를 없애는 함수
def turncate(a):
    if len(a) <= 3:
        return a
    elif 3 < len(a) <= 7:
        return a.split(',')[0] + a.split(',')[1]
    elif 7 < len(a) <= 11:
        return a.split(',')[0] + a.split(',')[1] + a.split(',')[2]
    else:
        return a
```

```python
# 숫자 사이의 ',' 제거
df['거래금액(만원)'] = df['거래금액(만원)'].apply(lambda x : x.strip())
df['거래금액(만원)'] = df['거래금액(만원)'].apply(turncate)
```

```python
# 데이터 타입 변경
df['전용면적(㎡)'] = df['전용면적(㎡)'].astype(float)
df['건축년도'] = df['건축년도'].astype(int)
df['거래금액(만원)'] = df['거래금액(만원)'].astype(int)
```

```python
# 종속 변수 : '면적당 가격' 만들기
df['면적당가격'] = df['거래금액(만원)'] / df['전용면적(㎡)']
```

```python
# 단위 삭제
df.rename(columns={'전용면적(㎡)':'전용면적',
                   '거래금액(만원)':'거래금액'}, inplace=True)
```

```python
# 지역 칼럼 분류
df['구'] = df['시군구'].apply(lambda x : x.split()[1])
df['동'] = df['시군구'].apply(lambda x : x.split()[2])
```

```python
# 불필요 칼럼 제거
df = df[['시군구', '전용면적', '거래금액', '층', '건축년도', '계약년월일', '면적당가격', '구', '동']]
```

```python
df.info()
```

```python
df['년월'] = df['계약년월일'].apply(lambda x : x[0:6])
```

```python
# 데이트타임 변환
df['계약년월일'] = pd.to_datetime(df['계약년월일'])
```

```python
# 데이트타임 분리
df['계약년'] = df['계약년월일'].dt.year
df['계약월'] = df['계약년월일'].dt.month
df['계약일'] = df['계약년월일'].dt.day
```
